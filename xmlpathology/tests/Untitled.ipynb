```python
import json
import numpy as np
from shapely.geometry import shape, GeometryCollection, Polygon
from matplotlib import pyplot as plt
import glob
from xmlpathology.xio.imageannotation import ImageAnnotation
from xmlpathology.xio.annotationparser import QuPathParser, AnnotationParserLoader
from xmlpathology.xio.jsontoxml import convert_image_annotation_to_xml
import os
```

```python
color_map = {'normal': '#800080', 'tls': '#FFFF00', 'gc': '#008000', 'lymph node': '#013220'}
```

```python
label_map = {'normal': 1, 'tls': 2, 'gc': 3, 'lymph node': 4}
```

```python
json_files  = glob.glob('/home/mart/jsonannotations/*.json')
```

```python
for json_file in json_files:
    imgannotation = ImageAnnotation(0, None, json_file, label_map=label_map, annotation_parser_loader=AnnotationParserLoader(_class=QuPathParser))
    output_path = json_file.replace('.json', '.xml')
    convert_image_annotation_to_xml(imgannotation, output_path, label_map, color_map)
    
```

```python
for annotation in imgannotation.annotations:    
    plt.plot(*annotation.exterior.xy)
plt.gca().invert_yaxis()
plt.axis('equal')
plt.show()
```
